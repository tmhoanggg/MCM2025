```python
import pandas as pd
import numpy as np

programs = pd.read_csv('../data/summerOly_programs-fixed.csv', index_col=False)
```

```python
year_cols = programs.columns[4:]

# Tính tổng sự kiện theo từng môn thể thao
discipline_sum_events = programs.groupby(['Sport', 'Discipline'])[year_cols].sum().sum(axis=1)

# Đổi tên cột kết quả thành 'discipline_sum_events'
discipline_sum_events = discipline_sum_events.reset_index(name='discipline_sum_events')
discipline_sum_events
```

```python
# Thêm cột discipline_sum_events vào df
programs['discipline_sum_events'] = discipline_sum_events['discipline_sum_events']
programs
```

```python
programs[(programs['1996'] == 0) & (programs['2016'] == 0) & (programs['2020'] == 0) & (programs['2024'] == 0)]
```

```python
# Lọc ra những discipline có số events là 0 ở 3 năm gần nhất (2016, 2020, 2024) và 0 ở năm 1996 (năm gần nhất Mỹ đăng cai) và bỏ đi chúng
condition = (programs['1996'] == 0) & (programs['2016'] == 0) & (programs['2020'] == 0) & (programs['2024'] == 0)
programs = programs.drop(programs[condition].index)
```

```python
# Drop 3 hàng cuối để tính lại
df = programs.drop(programs[-3:].index)
```

## Tính lại tổng số events, disciplines, sports
### Events

```python
# Lấy các cột từ năm 1896 đến 2024
year_columns = df.columns[4:]  # Bỏ qua các cột Sport, Discipline, Code, Sports Governing Body

# Tính tổng số sự kiện cho mỗi năm
total_events_per_year = df[year_columns].sum(axis=0)
```

```python
total_events_per_year
```

### Disciplines

```python
# Tính tổng số môn thi đấu có ít nhất một sự kiện trong mỗi năm
total_disciplines_per_year = (programs[year_columns] > 0).sum()
total_disciplines_per_year
```

### Sports

```python
# Tính tổng số môn thể thao duy nhất có ít nhất một sự kiện trong mỗi năm
total_sports_per_year = programs[year_columns].gt(0).groupby(programs['Sport']).any().sum()
total_sports_per_year
```

### Concat

```python
# Tạo hàng Total Events
total_events_row = pd.DataFrame([['Total events', 'Total events', 'Total events', 'Total events'] + total_events_per_year.tolist()], columns=df.columns)

# Tạo hàng Total Disciplines
total_disciplines_row = pd.DataFrame([['Total disciplines', 'Total disciplines', 'Total disciplines', 'Total disciplines'] + total_disciplines_per_year.tolist()], columns=df.columns)

# Tạo hàng Total Sports
total_sports_row = pd.DataFrame([['Total sports', 'Total sports', 'Total sports', 'Total sports'] + total_sports_per_year.tolist()], columns=df.columns)

# Thêm các hàng tổng hợp vào DataFrame đã lọc
final_df = pd.concat([df, total_events_row, total_disciplines_row, total_sports_row], ignore_index=True)
```
